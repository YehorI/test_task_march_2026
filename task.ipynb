# Реализация логики на python

Убрал все диагностики оставив только код выполняющий логику

```python
import pandas as pd
```

```python
checkout = pd.read_csv('store_checkout_queues.csv')
stores = pd.read_csv('store_stores.csv')

print(len(checkout), len(stores))
```

```python
mcheckout = pd.merge(
    checkout,
    stores,
    on='store_uuid'
)
```

```python
chosen_store_ids = [
    98451680,
    12864064,
]

chosen_stores = mcheckout.loc[
    (mcheckout['store_id'].isin(chosen_store_ids)), :
]
len(chosen_stores)
```

Сохраним результат в csv

```python
chosen_stores.to_csv('result.csv', index=False)
```

# Та же логика на SQL

Используем duckdb как базу данных позволяющую делать sql запросы напрямую к csv файлам

```python
import duckdb

query = f"""
    select
        scq.*,
        ss.store_id as store_id
    from
        store_checkout_queues.csv scq
    join
        store_stores.csv ss
        on scq.store_uuid = ss.store_uuid
    where
        store_id in {'(' + ','.join([str(i) for i in chosen_store_ids]) + ')'}
"""

duckdb.sql(query).write_csv('result2.csv')
```

```python
r1 = pd.read_csv('result.csv')
r2 = pd.read_csv('result2.csv')
```
